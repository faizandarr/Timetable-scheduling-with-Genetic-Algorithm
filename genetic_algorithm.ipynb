```python
import random

class Chromosome:
    def __init__(self, courses, classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes):
        self.course = courses
        self.class_type = classes[0]
        self.section = random.choice(sections)
        self.section_strength = random.choice(section_strengths)
        self.professor = random.choice(professors)
        self.first_lecture_day = random.choice(days)
        self.first_lecture_timeslot = random.choice(timeslots)
        self.first_lecture_room = self.second_lecture_room = random.choice(rooms)
        self.first_lecture_room_size = self.second_lecture_room_size= random.choice(room_sizes)
        self.second_lecture_day = random.choice(days)
        #self.second_lecture_timeslot = random.choice(timeslots)
        #self.second_lecture_room = random.choice(rooms)
        #self.second_lecture_room_size = random.choice(room_sizes)
        if self.course == "AI lab":
            self.second_lecture_timeslot = timeslots[-1]  # Set the last timeslot for labs
            self.first_lecture_timeslot = timeslots[-1]  # Set the last timeslot for labs
            self.class_type = 'Lab'

        else:
            self.second_lecture_timeslot = random.choice(timeslots)
            self.first_lecture_timeslot = random.choice(timeslots)


    def to_binary(self, courses, classes, sections, professors, days, timeslots, rooms, room_sizes):
        binary_representation = ''
        binary_representation += format(courses.index(self.course), '03b')
        binary_representation += '1' if self.class_type == 'Lab' else '0'
        binary_representation += format(sections.index(self.section), '03b')
        binary_representation += format(section_strengths.index(self.section_strength), '01b')
        binary_representation += format(professors.index(self.professor), '04b')
        binary_representation += format(days.index(self.first_lecture_day), '03b')
        binary_representation += format(timeslots.index(self.first_lecture_timeslot), '02b')
        binary_representation += format(rooms.index(self.first_lecture_room), '04b')
        binary_representation += format(room_sizes.index(self.first_lecture_room_size), '02b')
        binary_representation += format(days.index(self.second_lecture_day), '03b')
        binary_representation += format(timeslots.index(self.second_lecture_timeslot), '02b')
        binary_representation += format(rooms.index(self.second_lecture_room), '04b')
        binary_representation += format(room_sizes.index(self.second_lecture_room_size), '02b')
        return binary_representation

    def from_binary(binary_representation, courses, classes, sections, professors, days, timeslots, rooms, room_sizes):
        course_index = int(binary_representation[:4], 2)
        class_index = int(binary_representation[4:5], 2)
        section_index = int(binary_representation[5:8], 2)
        section_strength_index = int(binary_representation[8:10], 2)
        professor_index = int(binary_representation[10:14], 2)
        day_index = int(binary_representation[14:18], 2)
        first_lecture_timeslot_index = int(binary_representation[18:22], 2)
        first_lecture_room_index = int(binary_representation[22:26], 2)
        first_lecture_room_size_index = int(binary_representation[26:28], 2)

        # Convert indices to corresponding values
        course = courses[course_index]
        class_type = classes[class_index]
        section = sections[section_index]
        section_strength = [60, 120][section_strength_index]
        professor = professors[professor_index]
        first_lecture_day = days[day_index]
        first_lecture_timeslot = timeslots[first_lecture_timeslot_index]
        first_lecture_room = rooms[first_lecture_room_index]
        first_lecture_room_size = room_sizes[first_lecture_room_size_index]

        return Chromosome(course, class_type, section, section_strength, professor, first_lecture_day,
                   first_lecture_timeslot, first_lecture_room, first_lecture_room_size)

    @staticmethod
    def crossover(chromosome1, chromosome2):
        # Shuffle the attributes between chromosome1 and chromosome2
        child1 = Chromosome(
            chromosome1.course, [chromosome1.class_type,chromosome2.class_type], [chromosome1.section, chromosome2.section], section_strengths,
            [chromosome1.professor, chromosome2.professor], [chromosome1.first_lecture_day, chromosome2.first_lecture_day], [chromosome2.first_lecture_timeslot, chromosome1.first_lecture_timeslot],
            [chromosome1.first_lecture_room, chromosome2.first_lecture_room], [chromosome2.first_lecture_room_size, chromosome1.first_lecture_room_size]
        )

        child2 = Chromosome(
            chromosome1.course, [chromosome1.class_type,chromosome2.class_type], [chromosome1.section, chromosome2.section], section_strengths,
            [chromosome1.professor, chromosome2.professor], [chromosome1.first_lecture_day, chromosome2.first_lecture_day], [chromosome2.first_lecture_timeslot, chromosome1.first_lecture_timeslot],
            [chromosome1.first_lecture_room, chromosome2.first_lecture_room], [chromosome2.first_lecture_room_size, chromosome1.first_lecture_room_size]
        )

        return child1, child2

    @staticmethod
    def mutation(chromosome):
      random_key1 = random.randint(1, 5)
      if random_key1 == 1:
        chromosomes = Chromosome(
          chromosome.course, [chromosome.class_type,chromosome.class_type], sections, section_strengths,
          [chromosome.professor, chromosome.professor], [chromosome.first_lecture_day, chromosome.first_lecture_day], [chromosome.first_lecture_timeslot, chromosome.first_lecture_timeslot],
          [chromosome.first_lecture_room, chromosome.first_lecture_room], [chromosome.first_lecture_room_size, chromosome.first_lecture_room_size]
        )
      elif random_key1 ==2:
        chromosomes = Chromosome(
          chromosome.course, [chromosome.class_type,chromosome.class_type], sections, section_strengths,
          professors, [chromosome.first_lecture_day, chromosome.first_lecture_day], [chromosome.first_lecture_timeslot, chromosome.first_lecture_timeslot],
          [chromosome.first_lecture_room, chromosome.first_lecture_room], [chromosome.first_lecture_room_size, chromosome.first_lecture_room_size]
        )
      return

    def __str__(self):
        return f"Course: {self.course}, Class Type: {self.class_type}, Section: {self.section}, Section Strength: {self.section_strength}, Professor: {self.professor}, First Lecture Day: {self.first_lecture_day}, First Lecture Timeslot: {self.first_lecture_timeslot}, First Lecture Room: {self.first_lecture_room}, First Lecture Room Size: {self.first_lecture_room_size}, Second Lecture Day: {self.second_lecture_day}, Second Lecture Timeslot: {self.second_lecture_timeslot}, Second Lecture Room: {self.second_lecture_room}, Second Lecture Room Size: {self.second_lecture_room_size}"

# Example usage:
courses = ["PDC", "SE", "AI", "NC", "SMD",'AI lab']
classes = ["Class", "Lab"]
sections = ["A", "B", "C", "D", "E"]
section_strengths = [60, 120]
professors = ["Mr. Smith", "Ms. Johnson", "Mr. Brown", "Dr. White", "Mrs. Davis", "Ms. Martinez", "Mr. Lee", "Mrs. Clark", "Dr. Taylor", "Ms. Hall"]
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
timeslots = ["8:30 AM - 10:00 AM", "10:15 AM - 12:45 PM", "1:00 PM - 2:30 PM", "2:45 PM - 5:00 PM"]
rooms = ["C301", "C302", "C303", "C304", "C305", "C401", "C402", "C403", "C404", "C405"]
room_sizes = ["Small", "Medium", "Large"]


chromosomes_AI = []
chromosomes_AI_lab = []
chromosomes_PDC = []
chromosomes_NC = []
chromosomes_SMD = []

chromosomes_AI_BN = []
chromosomes_AI_lab_BN = []
chromosomes_PDC_BN = []
chromosomes_NC_BN = []
chromosomes_SMD_BN = []

'''for i in range(5):
    chromosomes_AI.append(Chromosome('AI', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes))
    chromosomes_AI_BN.append(Chromosome('AI', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes).to_binary(courses, classes, sections, professors, days, timeslots, rooms, room_sizes))
    chromosomes_AI_lab.append(Chromosome('AI lab', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes))
    chromosomes_AI_lab_BN.append(Chromosome('AI lab', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes).to_binary(courses, classes, sections, professors, days, timeslots, rooms, room_sizes))
    chromosomes_PDC.append(Chromosome('PDC', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes))
    chromosomes_PDC_BN.append(Chromosome('PDC', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes).to_binary(courses, classes, sections, professors, days, timeslots, rooms, room_sizes))
    chromosomes_NC.append(Chromosome('NC', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes))
    chromosomes_NC_BN.append(Chromosome('NC', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes).to_binary(courses, classes, sections, professors, days, timeslots, rooms, room_sizes))
    chromosomes_SMD.append(Chromosome('SMD', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes))
    chromosomes_SMD_BN.append(Chromosome('SMD', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes).to_binary(courses, classes, sections, professors, days, timeslots, rooms, room_sizes))





Total_chromosomes = []
Total_chromosomes.append(chromosomes_AI)
Total_chromosomes.append(chromosomes_AI_lab)
Total_chromosomes.append(chromosomes_PDC)
Total_chromosomes.append(chromosomes_NC)
Total_chromosomes.append(chromosomes_SMD)

for chromosome in Total_chromosomes:
  for chromo in chromosome:
    print(chromo)
'''
###############################################################################################
#FITNESS CHECKING
def hard_constraints(chromosomes):
    conflict = 0

    # Dictionary to keep track of occupied rooms and timeslots
    occupied_rooms = {}
    occupied_timeslots = {}
    professor_courses = {}
    section_courses = {}
    course_lectures = {}
    course_classroom = {}
    course_timeslot = {}
    room_floor = {}
    previous_timeslot = {}

    # Iterate over each chromosome in the list of chromosomes
    for chromosome in chromosomes:
        # Constraint: Classes can only be scheduled in free classrooms.
        if (chromosome.first_lecture_day, chromosome.first_lecture_timeslot, chromosome.first_lecture_room) in occupied_rooms:
            conflict += 1
        else:
            occupied_rooms[(chromosome.first_lecture_day, chromosome.first_lecture_timeslot, chromosome.first_lecture_room)] = chromosome

        # Constraint: A classroom should be big enough to accommodate the section
        if chromosome.first_lecture_room_size == "Small" and chromosome.section_strength > 60:
            conflict += 1

        # Constraint: A professor should not be assigned two different lectures at the same time.
        if (chromosome.first_lecture_day, chromosome.first_lecture_timeslot) in occupied_timeslots:
            if occupied_timeslots[(chromosome.first_lecture_day, chromosome.first_lecture_timeslot)].professor == chromosome.professor:
                conflict += 1
            else:
                professor_courses[chromosome.professor] = professor_courses.get(chromosome.professor, 0) + 1
                if professor_courses[chromosome.professor] > 3:
                    conflict += 1
        else:
            occupied_timeslots[(chromosome.first_lecture_day, chromosome.first_lecture_timeslot)] = chromosome

        # Constraint: The same section cannot be assigned to two different rooms at the same time.
        if (chromosome.first_lecture_day, chromosome.first_lecture_timeslot, chromosome.section) in section_courses:
            if section_courses[(chromosome.first_lecture_day, chromosome.first_lecture_timeslot, chromosome.section)] != chromosome.first_lecture_room:
                conflict += 1
        else:
            section_courses[(chromosome.first_lecture_day, chromosome.first_lecture_timeslot, chromosome.section)] = chromosome.first_lecture_room

        # Constraint: A room cannot be assigned for two different sections at the same time.
        if (chromosome.first_lecture_day, chromosome.first_lecture_timeslot, chromosome.first_lecture_room) in occupied_rooms:
            if occupied_rooms[(chromosome.first_lecture_day, chromosome.first_lecture_timeslot, chromosome.first_lecture_room)].section != chromosome.section:
                conflict += 1
        else:
            occupied_rooms[(chromosome.first_lecture_day, chromosome.first_lecture_timeslot, chromosome.first_lecture_room)] = chromosome

        # Constraint: No section can have more than 5 courses in a semester.
        section_courses[chromosome.section] = section_courses.get(chromosome.section, 0) + 1
        if section_courses[chromosome.section] > 5:
            conflict += 1

        # Constraint: Each course would have two lectures per week not on the same or adjacent days.
        course_lectures[chromosome.course] = course_lectures.get(chromosome.course, [])
        course_lectures[chromosome.course].append(chromosome.first_lecture_day)
        if len(set(course_lectures[chromosome.course])) < 2:
            conflict += 1
        elif days.index(chromosome.first_lecture_day) + 1 == days.index(chromosome.second_lecture_day):
            conflict += 1
        elif days.index(chromosome.first_lecture_day) - 1 == days.index(chromosome.second_lecture_day):
            conflict += 1

        # Constraint: Lab lectures should be conducted in two consecutive slots.
        if chromosome.class_type == "Lab":
            if timeslots.index(chromosome.first_lecture_timeslot) + 1 != timeslots.index(chromosome.second_lecture_timeslot):
                conflict += 1

        # Soft Constraint: All theory classes should be taught in the morning session
        if chromosome.class_type == "Class" and not (timeslots.index(chromosome.first_lecture_timeslot) < 2):
            conflict += 0.5

        # Soft Constraint: All lab sessions should be done in the afternoon session
        if chromosome.class_type == "Lab" and not (timeslots.index(chromosome.first_lecture_timeslot) >= 2):
            conflict += 0.5

        # Soft Constraint: Teachers/students may be facilitated by minimizing the number of floors they have to traverse
        room_floor[chromosome.first_lecture_room] = room_floor.get(chromosome.first_lecture_room, 0)
        room_floor[chromosome.first_lecture_room] = max(room_floor[chromosome.first_lecture_room], rooms.index(chromosome.first_lecture_room) // 5)

        # Soft Constraint: A class should be held in the same classroom across the whole week
        if chromosome.course in course_classroom:
            if course_classroom[chromosome.course] != chromosome.first_lecture_room:
                conflict += 0.5
        else:
            course_classroom[chromosome.course] = chromosome.first_lecture_room

        # Soft Constraint: Teachers may prefer longer blocks of continuous teaching time to minimize interruptions
        if chromosome.professor in previous_timeslot:
            if timeslots.index(chromosome.first_lecture_timeslot) - timeslots.index(previous_timeslot[chromosome.professor]) > 1:
                conflict += 0.5
        previous_timeslot[chromosome.professor] = chromosome.first_lecture_timeslot

    return conflict

########################################      SELECTION FUNCTION       #########################################################################
def Selection(Total_chromosomes):
    for course_chromosomes in Total_chromosomes:
        conflicts = hard_constraints(course_chromosomes)
        sorted_chromosomes = sorted(course_chromosomes, key=lambda chromo: hard_constraints([chromo] + [other_chromo for other_course_chromosomes in Total_chromosomes for other_chromo in other_course_chromosomes if other_chromo != chromo]))

        print(f"Number of conflicts for {course_chromosomes[0].course}: {conflicts}")
        for chromo in sorted_chromosomes:
            print(chromo)
    return Total_chromosomes

########################################       CROSSOVER FUNCTION        #########################################################
def CrossOver(Course_chromosome):
  random_val1 = 0
  random_val2 = 1


  parent1 = Course_chromosome[random_val1]
  parent2 = Course_chromosome[random_val2]
  child1, child2 = Chromosome.crossover(parent1, parent2)

  Course_chromosome[random_val1]=child1
  Course_chromosome[random_val2]=child2
  return Course_chromosome

 ############################################  MUTATION FUNCTION #######################################################
def Mutation(Course_chromosome):
  for chromo in Course_chromosome:
    Chromosome.mutation(chromo)
  return Course_chromosome

###############################################################################################################################
'''print("\nChromosomes for AI:")
for chromosome, chromosome_BN in zip(chromosomes_AI, chromosomes_AI_BN):
    print(chromosome)
    print(chromosome_BN)
print("\nChromosomes for AI-lab:")
for chromosome, chromosome_BN in zip(chromosomes_AI_lab, chromosomes_AI_lab_BN):
    #print(chromosome)
    print(chromosome_BN)
print("\nChromosomes for PDC:")
for chromosome, chromosome_BN in zip(chromosomes_PDC, chromosomes_PDC_BN):
    #print(chromosome)
    print(chromosome_BN)
print("\nChromosomes for NC:")
for chromosome, chromosome_BN in zip(chromosomes_NC, chromosomes_NC_BN):
    #print(chromosome)
    print(chromosome_BN)
print("\nChromosomes for SMD:")
for chromosome, chromosome_BN in zip(chromosomes_SMD, chromosomes_SMD_BN):
    #print(chromosome)
    print(chromosome_BN)

TOTAL_HARD_CONSTRAINTS_CONFLICTS = 0

conflicts = hard_constraints(chromosomes_AI)
print("Number of conflicts of AI:", conflicts)
TOTAL_HARD_CONSTRAINTS_CONFLICTS += conflicts

conflicts = hard_constraints(chromosomes_AI_lab)
print("Number of conflicts of AI-lab:", conflicts)
TOTAL_HARD_CONSTRAINTS_CONFLICTS += conflicts

conflicts = hard_constraints(chromosomes_PDC)
print("Number of conflicts of PDC:", conflicts)
TOTAL_HARD_CONSTRAINTS_CONFLICTS += conflicts

conflicts = hard_constraints(chromosomes_NC)
print("Number of conflicts of NC:", conflicts)
TOTAL_HARD_CONSTRAINTS_CONFLICTS += conflicts

conflicts = hard_constraints(chromosomes_SMD)
print("Number of conflicts of SMD:", conflicts)
TOTAL_HARD_CONSTRAINTS_CONFLICTS += conflicts

print('\n\n\nTOTAL HARD CONSTRAINTS CONFLICTS = ',TOTAL_HARD_CONSTRAINTS_CONFLICTS)
Selection(Total_chromosomes)
'''

##########################################GENETIC ALGORITHM #####################################################################

def GeneticAlgorithm():
  print("########################################GENETIC ALGORITHM:########################################################3")
  generation_limit = 50
  for i in range(5):
    chromosomes_AI.append(Chromosome('AI', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes))
    chromosomes_AI_BN.append(Chromosome('AI', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes).to_binary(courses, classes, sections, professors, days, timeslots, rooms, room_sizes))
    chromosomes_AI_lab.append(Chromosome('AI lab', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes))
    chromosomes_AI_lab_BN.append(Chromosome('AI lab', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes).to_binary(courses, classes, sections, professors, days, timeslots, rooms, room_sizes))
    chromosomes_PDC.append(Chromosome('PDC', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes))
    chromosomes_PDC_BN.append(Chromosome('PDC', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes).to_binary(courses, classes, sections, professors, days, timeslots, rooms, room_sizes))
    chromosomes_NC.append(Chromosome('NC', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes))
    chromosomes_NC_BN.append(Chromosome('NC', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes).to_binary(courses, classes, sections, professors, days, timeslots, rooms, room_sizes))
    chromosomes_SMD.append(Chromosome('SMD', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes))
    chromosomes_SMD_BN.append(Chromosome('SMD', classes, sections, section_strengths, professors, days, timeslots, rooms, room_sizes).to_binary(courses, classes, sections, professors, days, timeslots, rooms, room_sizes))

  Total_chromosomes = []
  Total_chromosomes.append(chromosomes_AI)
  Total_chromosomes.append(chromosomes_AI_lab)
  Total_chromosomes.append(chromosomes_PDC)
  Total_chromosomes.append(chromosomes_NC)
  Total_chromosomes.append(chromosomes_SMD)


  for i in range(generation_limit):
    print("||||||||||||||||||||||||||GENERATION COUNT: ",i,'||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||')


    Total_chromosomes = Selection(Total_chromosomes)
    CrossOver(chromosomes_AI)
    CrossOver(chromosomes_AI_lab)
    CrossOver(chromosomes_PDC)
    CrossOver(chromosomes_NC)
    CrossOver(chromosomes_SMD)
    Mutation(chromosomes_AI)
    Mutation(chromosomes_AI_lab)
    Mutation(chromosomes_PDC)
    Mutation(chromosomes_NC)
    Mutation(chromosomes_SMD)

    TOTAL_HARD_CONSTRAINTS_CONFLICTS = 0

    conflicts = hard_constraints(chromosomes_AI)
    print("Number of conflicts of AI:", conflicts)
    TOTAL_HARD_CONSTRAINTS_CONFLICTS += conflicts

    conflicts = hard_constraints(chromosomes_AI_lab)
    print("Number of conflicts of AI-lab:", conflicts)
    TOTAL_HARD_CONSTRAINTS_CONFLICTS += conflicts

    conflicts = hard_constraints(chromosomes_PDC)
    print("Number of conflicts of PDC:", conflicts)
    TOTAL_HARD_CONSTRAINTS_CONFLICTS += conflicts

    conflicts = hard_constraints(chromosomes_NC)
    print("Number of conflicts of NC:", conflicts)
    TOTAL_HARD_CONSTRAINTS_CONFLICTS += conflicts

    conflicts = hard_constraints(chromosomes_SMD)
    print("Number of conflicts of SMD:", conflicts)
    TOTAL_HARD_CONSTRAINTS_CONFLICTS += conflicts


    print('Total Conflicts: ',TOTAL_HARD_CONSTRAINTS_CONFLICTS )


GeneticAlgorithm()

```
